```python
import matplotlib.pyplot as plt

# Sample data (adjust to match your original figure)
emotions = ["Fear", "Anger", "Sadness", "Anticipation", "Joy", "Trust", "Disgust", "Surprise"]
one_month = [31, 19, 16, 14, 13, 6, 10, 6]
mask_wearing = [16, 11, 14, 22, 7, 14, 8, 6]
lockdown = [23, 17, 14, 15, 12, 6, 7, 6]

# Choose a color-blind-friendly palette
colors = ["#007bff", "#ff7f0e", "#2ca02c"]  # Blue, Orange, Green

# Create bar chart with distinct colors
bar_width = 0.3  # Increased bar width
spacing = 0.1    # Space between groups of bars

# Adjust x positions for each set of bars
x_positions_one_month = range(len(emotions))
x_positions_mask_wearing = [x + bar_width + spacing for x in x_positions_one_month]
x_positions_lockdown = [x + 2 * (bar_width + spacing) for x in x_positions_one_month]

plt.bar(x_positions_one_month, one_month, color=colors[0], label="One Month", width=bar_width)
plt.bar(x_positions_mask_wearing, mask_wearing, color=colors[1], label="Mask-wearing Period", width=bar_width)
plt.bar(x_positions_lockdown, lockdown, color=colors[2], label="Lockdown Period", width=bar_width)

# Set title and axes labels
plt.title("Emotion Distribution Analysis Across Different Periods")
plt.xlabel("Emotion")
plt.ylabel("Percentage")

# Add numbers to each bar
for i in range(len(emotions)):
    plt.text(i, one_month[i] + 1, str(one_month[i]), ha='center', va='bottom')
    plt.text(x_positions_mask_wearing[i], mask_wearing[i] + 1, str(mask_wearing[i]), ha='center', va='bottom')
    plt.text(x_positions_lockdown[i], lockdown[i] + 1, str(lockdown[i]), ha='center', va='bottom')

# Customize x-ticks
plt.xticks([x + (bar_width + spacing) for x in range(len(emotions))], emotions)

# Customize legend and add context note
plt.legend()
plt.figtext(0.5, 0.01, "Note: Total tweets analyzed (approximately): One Month: 1000, Mask-wearing Period: 800, Lockdown Period: 750", ha="center", fontsize=8)

# Adjust layout and display the chart
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for one emotion (e.g., "Fear")
labels = ["One Month", "Mask-wearing Period", "Lockdown Period"]
sizes = [31, 16, 23]  # Corresponding values for "Fear"

# Choose a color-blind-friendly palette
colors = ["#007bff", "#ff7f0e", "#2ca02c"]  # Blue, Orange, Green

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of 'Fear' Emotion Across Different Periods")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the chart
plt.tight_layout()
plt.show()
```
